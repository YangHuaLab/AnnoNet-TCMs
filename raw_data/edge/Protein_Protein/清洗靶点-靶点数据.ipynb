```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Protein_Protein_in_STRING.txt",sep=' ')
protein_info = pd.read_csv(r"D:\Herb数据库\version2\node\Protein_info_in_STRING.txt",sep='\t')[['#string_protein_id','preferred_name']]
protein_info.rename(columns={'#string_protein_id':'protein','preferred_name':'Gene_name'},inplace=True)
```

```python
data_with_name = pd.merge(
    pd.merge(data,
            protein_info[['Gene_name','protein']],left_on='protein1',right_on='protein',how='left').iloc[:,:-1],
    protein_info[['Gene_name','protein']],left_on='protein2',right_on='protein',how='left').iloc[:,:-1]
data_with_name.rename(columns={'Gene_name_x':'Gene_name1','Gene_name_y':'Gene_name2'},inplace=True)
data_exp = data_with_name.query("experimental>0")
```

```python
len(set(data_exp['Gene_name1'])),len(set(data_exp['Gene_name2']))
```

```python
len(set(data_exp['protein1'])),len(set(data_exp['protein2']))
```

```python
all_target = list(set(data_exp['Gene_name1'])|set(data_exp['Gene_name2']))
```

```python
protein_info_filtered = pd.read_csv(r"D:\Herb数据库\version2\node\Protein_info_in_STRING.txt",sep='\t').query("preferred_name in @all_target")
protein_info_filtered.rename(columns={'#string_protein_id':'protein','preferred_name':'Gene_name'},inplace=True)
```

```python
protein_info_filtered.to_csv(r"D:\Herb数据库\version2\node\Target_info_in_physical_network_and_experimental.csv")
```
