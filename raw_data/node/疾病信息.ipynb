```python
import pandas as pd
disease_in_nw = pd.read_excel("Disease_info_with_cn_name.xlsx",index_col=0)
disease_all = pd.read_csv("Disease_info_in_DisGENet.csv",index_col=0)
```

```python
def add_string(x):
    string = ''
    for i,j in enumerate(x):
        if i==0:
            string += j
        else:
            string = string + '; '+ j
    return string
```

```python
disease_in_nw_with_info = pd.merge(disease_in_nw,disease_all,on='Disease_name',how='left')
disease_in_nw_with_info['Disease_class_name'] = disease_in_nw_with_info['Disease_class_name'].apply(lambda x: str(x).strip())
disease_in_nw_with_info.groupby(['Disease_name','Disease_type'])['Disease_class_name'].apply(add_string).to_excel('Disease_info_all.xlsx')
```
