```python
import pandas as pd
import numpy as np
from rdkit import Chem
from tqdm import tqdm
```

```python
data = pd.read_csv('chemicals.v5.0.tsv',sep='\t')
```

```python
comp_targ = pd.read_csv(r"D:\Herb数据库\连接表\9606.protein_chemical.links.detailed.v5.0.tsv",sep='\t')
```

```python
have_link_comp = list(set(comp_targ['chemical']))
```

```python
have_link_data = data.query("chemical in @have_link_comp")
```

```python
can_smi = []
for smi in tqdm(have_link_data['SMILES_string']):
    try:
        can_smi.append(Chem.MolToSmiles(Chem.MolFromSmiles(smi),kekuleSmiles=True,isomericSmiles=False))
    except:
        can_smi.append(None)
```

```python
have_link_data['Canonical_smiles'] = can_smi
```

```python
can_smiles = list(set(pd.read_csv(r"D:\Herb数据库\药材-成分数据\成分Canonical Smiles数据0130.csv",index_col=0)['Canonical_smiles']))
```

```python
useful_comp_data = have_link_data.query("Canonical_smiles in @can_smiles")
```

```python
useful_comp_data.to_excel("有靶点连接以及有药材连接的成分0201.xlsx")
```
